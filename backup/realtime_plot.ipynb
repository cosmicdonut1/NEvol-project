```python
import numpy as np
from datetime import datetime, timedelta

# Generate sample data (500, 8)
sample_data = np.random.rand(500, 8)

# Generate timestamps for the data
start_time = datetime.now()
lsl_timestamps = np.array([start_time + timedelta(seconds=i) for i in range(500)])

# Example usage:
print("Sample Data shape:", sample_data.shape)
print("Timestamps length:", lsl_timestamps.shape)
```

```python
import matplotlib.pyplot as plt

# Plotting the data
fig, axs = plt.subplots(8, figsize=(12, 16), sharex=True)

for i in range(8):
    axs[i].plot(lsl_timestamps, sample_data[:, i])
    axs[i].set_title(f'Data Series {i+1}')

# Labeling and formatting
plt.xlabel('Timestamp')
plt.tight_layout()
plt.show()
```
